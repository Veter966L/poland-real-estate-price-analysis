```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pol_apart = pd.read_csv('datasets/archive1/apartments_pl_2024_01.csv') #Download dataset as pol_apart
```

```python
print(pol_apart.shape) #Dataset consisits of 28 columns and 15521 strings
```

```python
print(pol_apart.info()) #There are passes in coolumns
missing_pol_apart = pol_apart.isnull().sum().sort_values(ascending=False)
missing_pol_apart_perc = missing_pol_apart / len(pol_apart) * 100
a = 20
print(pd.DataFrame({
    'Missing': missing_pol_apart,
    'Percent': missing_pol_apart_perc})) # 'condition': 76% and 'buildingMaterial':38% - critical missings
                                         # 'type':22%, 'floor':18% and 'buildYear':16% - intermediate missings

    
```

```python
#Exploring: how much missed values impact for price
pol_apart['condition'] = pol_apart['condition'].fillna('unknown') #change missing to 'unknown' category
print(pol_apart['condition'].unique())
condition_price = pol_apart.groupby('condition')['price'].describe()
print(condition_price) #Mean value of "unknown" category closer to "premium". In all categories there is a huge variation - the price depends not only
                        #on the condition. Minimal and maximum values "price" have category "unknown"
                        #Suggestion: Apartment's owners forget to fill in field "condition"
```

```python
year_condition = pol_apart.groupby('condition')['buildYear'].describe() # Most apartment 
print(year_condition)
sns.boxplot(data=pol_apart, x='condition', y='price')
plt.show()
```

```python
# Двойной график: цена и год постройки по condition
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# График 1: Средняя цена
avg_price = pol_apart.groupby('condition')['price'].mean().sort_values(ascending=False)
colors_price = ['#2ecc71', '#95a5a6', '#e74c3c']  # зеленый, серый, красный

bars1 = axes[0].bar(avg_price.index, avg_price.values, color=colors_price, edgecolor='black', linewidth=1.5)
axes[0].set_title('Средняя цена по состоянию квартиры', fontweight='bold', fontsize=14)
axes[0].set_ylabel('Цена (PLN)', fontsize=12)
axes[0].set_xlabel('Состояние', fontsize=12)
axes[0].grid(axis='y', alpha=0.3)

# Добавляем значения на столбцы
for bar in bars1:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height/1000)}k',
                ha='center', va='bottom', fontweight='bold', fontsize=11)

# График 2: Средний год постройки
avg_year = pol_apart.groupby('condition')['buildYear'].mean().sort_values(ascending=False)
colors_year = ['#3498db', '#95a5a6', '#e67e22']  # синий, серый, оранжевый

bars2 = axes[1].bar(avg_year.index, avg_year.values, color=colors_year, edgecolor='black', linewidth=1.5)
axes[1].set_title('Средний год постройки по состоянию', fontweight='bold', fontsize=14)
axes[1].set_ylabel('Год постройки', fontsize=12)
axes[1].set_xlabel('Состояние', fontsize=12)
axes[1].set_ylim(1950, 2025)
axes[1].grid(axis='y', alpha=0.3)

# Добавляем значения
for bar in bars2:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height)}',
                ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.savefig('condition_price_year_comparison.png', dpi=300, bbox_inches='tight')
print("✓ График сохранен: condition_price_year_comparison.png")
plt.show()
```
